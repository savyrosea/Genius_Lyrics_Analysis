# 2. Set Up Empty DF's to Append To

```python
# appending features to artist df

artist_info = {
    'artist' : [''],
    'description' : [''],
    'images_url' : [''],
    'followers_count' : ['']
    }

artist_df = pd.DataFrame(artist_info)

song_info = {
    'song_name' : [''],
    'artist' : [''],
    'lyrics' : [''],
    'description' : [''],
    'accepted_annotations' : [''],
    'contributors' : [''],
    'pageviews' : ['']
        }

songs_df = pd.DataFrame(song_info)
```

# 3. Define a Function That Will Add Top 5 Songs For a List of Artists

```python
# This function takes a list of artists and adds them to my df's
def add_artists(list_artists):
    
    # global so it can use defined variables above
    global artist_df, songs_df
    
    # looping through all artists
    for art in list_artists:
        # read in top 5 songs for the artist
        artist = genius.search_artist(art, max_songs = 5)
        # save lyrics and song info to json
        artist.save_lyrics()
        # making json file path as string
        temp_art = art.replace(" ", "")
        temp_art = "Lyrics_" + temp_art + ".json"
        # reading in json
        with open(temp_art) as json_data:
            data = json.load(json_data)

        # appending features to artist df
        artist_info_temp = {
        'artist' : [data['name']],
        'description' : [data['description']['plain'].replace('\n',' ').replace("\'","'")],
        'images_url' : [data['image_url']],
        'followers_count' : [data['followers_count']]
        }
        artist_df_temp = pd.DataFrame(artist_info_temp)
        artist_df = artist_df.append(artist_df_temp, ignore_index=True)

        # appending features to song df
        for i in range(5):
            song_info_temp = {
            'song_name' : [data['songs'][i]['title']],
            'artist' : [data['songs'][i]['artist']],
            'lyrics' : [data['songs'][i]['lyrics'].replace('\n',' ').replace("\'","'")],
            'description' : [data['songs'][i]['description']['plain'].replace('\n',' ').replace("\'","'")],
            'accepted_annotations' : [data['songs'][i]['stats']['accepted_annotations']],
            'contributors' : [data['songs'][i]['stats']['contributors']],
            'pageviews' : [data['songs'][i]['stats']['pageviews']]
            }
            songs_df_temp = pd.DataFrame(song_info_temp)
            songs_df = songs_df.append(songs_df_temp, ignore_index=True)
```
